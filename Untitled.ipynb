```python
import numpy as np

def input_array_a():
    a = np.array([float(input(f"Введите элемент a[{i}]: ")) for i in range(10)])
    return a

def create_array_b():
    b = np.linspace(0, 9, 10)
    return b

def create_random_array_c():
    c = np.random.uniform(1, 5, 10)
    return c

def print_array_info(arr, name):
    print(f"Массив {name}:")
    print(arr)
    print("Shape:", arr.shape)
    print("Size:", arr.size)
    print("Ndim:", arr.ndim)
    print("Dtype:", arr.dtype)
    print()

def sum_of_minimals(a, b, c):
    return np.min(a) + np.min(b) + np.min(c)

def add_arrays(a, b):
    return a + b

def remove_even_positions(a):
    return a[1::2]

def insert_sequence(b):
    return np.insert(b, 3, [1, 1, 1])

def copy_array_a(a):
    d = np.copy(a)
    return d

def concatenate_arrays(a, b):
    return np.concatenate((a, b))

def split_array(d):
    half = len(d) // 2
    return d[:half], d[half:]

def reshape_array(a):
    return a.reshape(2, 5)

a = input_array_a()

b = create_array_b()
c = create_random_array_c()

print_array_info(a, "a")
print_array_info(b, "b")
print_array_info(c, "c")

result1 = sum_of_minimals(a, b, c)
result2 = add_arrays(a, b)
result3 = remove_even_positions(a)
result4 = insert_sequence(b)
d = copy_array_a(a)
result5 = concatenate_arrays(a, b)
result6, result7 = split_array(d)
result8 = reshape_array(a)

print("\nРезультаты задач:")
print("Сумма минимальных элементов каждого массива:", result1)
print("Сложение массивов a и b:")
print(result2)
print("Массив a после удаления чётных позиций:")
print(result3)
print("Массив b после вставки [1,1,1] между 3 и 4 позициями:")
print(result4)
print("Результат объединения массивов a и b:")
print(result5)
print("Первая половина массива d:")
print(result6)
print("Вторая половина массива d:")
print(result7)
print("Двумерный массив из a:")
print(result8)
```

```python
import numpy as np

# Функция для ввода массива a с клавиатуры
def input_array_a():
    a = np.empty((4, 4))
    for i in range(4):
        for j in range(4):
            a[i, j] = float(input(f"Введите элемент a[{i}][{j}]: "))
    return a


# Функция для формирования массива b с произвольными числами
def create_array_b():
    b = np.random.uniform(-10, 10, (4, 4))
    return b

# Функция для вывода массива и его атрибутов
def print_array_info(arr, name):
    print(f"Массив {name}:")
    print(arr)
    print("Shape:", arr.shape)
    print("Size:", arr.size)
    print("Ndim:", arr.ndim)
    print("Dtype:", arr.dtype)
    print()

# Функция для нахождения минимального и максимального элементов в каждой строке массива a
# и поменять их местами
def swap_min_max_elements(a):
    for i in range(4):
        min_idx = np.argmin(a[i])
        max_idx = np.argmax(a[i])
        a[i][min_idx], a[i][max_idx] = a[i][max_idx], a[i][min_idx]
    return a

# Функция для нахождения среднего арифметического чётных элементов третьего столбца массива b
def average_of_even_elements(b):
    third_column = b[:, 2]
    even_elements = third_column[third_column % 2 == 0]
    return np.mean(even_elements)

# Функция для формирования одномерного массива из отрицательных элементов массива a
def negative_elements_of_a(a):
    return a[a < 0]

# Функция для возведения элементов второй строки массива a в квадрат
def square_second_row(a):
    a[1, :] = np.square(a[1, :])
    return a

# Функция для формирования и вывода нового массива из первого и последнего столбцов
# массива a, а затем преобразования его в одномерный
def extract_and_flatten_columns(a):
    new_array = a[:, [0, -1]]
    flattened_array = new_array.flatten()
    print("Новый массив из первого и последнего столбцов a:")
    print(new_array)
    print("Преобразованный в одномерный:")
    print(flattened_array)
    return flattened_array

# Функция для вывода части массива b от элемента b[1,3] до элемента b[3,3]
def print_part_of_b(b):
    part_b = b[1:4, 3]
    print("Часть массива b от элемента b[1,3] до элемента b[3,3]:")
    print(part_b)

# Функция для добавления новой строки между 2 и 3 строками массива a
def insert_new_row(a):
    # Создаем новую строку с значениями [10, 10, 10, 10]
    new_row = np.array([[10, 10, 10, 10]])
    # Вставляем новую строку между 2 и 3 строками (по индексу 2) массива a
    a = np.insert(a, 2, new_row, axis=0)
    return a

# Функция для транспонирования массива b
def transpose_b(b):
    transposed_b = np.transpose(b)
    return transposed_b

# Ввод массива a с клавиатуры
a = input_array_a()

# Формирование массива b с произвольными числами
b = create_array_b()

# Вывод информации о каждом массиве
print_array_info(a, "a")
print_array_info(b, "b")

# Выполнение задач
a = swap_min_max_elements(a)
average = average_of_even_elements(b)
negatives = negative_elements_of_a(a)
a = square_second_row(a)
flattened_a = extract_and_flatten_columns(a)
print_part_of_b(b)
a = insert_new_row(a)
transposed_b = transpose_b(b)

# Вывод результатов
print("Массив a после замены минимальных и максимальных элементов:")
print(a)
print("Среднее арифметическое чётных элементов третьего столбца b:", average)
print("Одномерный массив из отрицательных элементов a:", negatives)
print("Массив a после возведения второй строки в квадрат:")
print(a)
print("Преобразованный массив из первого и последнего столбцов a:")
print(flattened_a)
print("Массив b после транспонирования:")
print(transposed_b)
```
